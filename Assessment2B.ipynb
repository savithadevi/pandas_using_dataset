###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


###### if  10000<codekata score<15000   (Reached_expectations.csv)



###### if  7000<codekata score<10000   (Needs_Improvement.csv)


###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
data = pd.read_csv("/content/college_1.csv")
 

Exceeded_expectations = data[data['CodeKata Score'] >=15000]
Reached_expectations = data[data['CodeKata Score'] >=10000]
Needs_Improvement = data[data['CodeKata Score'] <10000]
Unsatisfactory = data[data['CodeKata Score'] <7000]

Exceeded_expectations.to_csv('Exceeded expectations.csv', index=False) 
Reached_expectations.to_csv('Reached_expectations.csv', index=False)
Needs_Improvement.to_csv('Needs_Improvement.csv', index=False)
Unsatisfactory.to_csv('Unsatisfactory.csv', index=False)
 
(pd.read_csv("Exceeded expectations.csv"))
(pd.read_csv("Reached_expectations.csv"))
(pd.read_csv("Needs_Improvement.csv"))
(pd.read_csv("Unsatisfactory.csv"))
```

###### Department wise codekata performence (pie chart)

```python

import matplotlib.pyplot as plt
import plotly.express as px
df = pd.concat(
    map(pd.read_csv,['/content/college_1.csv', '/content/college_2.csv']), ignore_index=True)

fig=px.pie(df,values ='CodeKata Score',names='Department')
fig.show()
```
